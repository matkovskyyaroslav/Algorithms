```python
import sys

sys.setrecursionlimit(1000000)


class Graph:
    def __init__(self, adj_list):
        self.adj_list = adj_list
        self.white = 0
        self.gray = 1
        self.black = 2
        self.colors = [self.white] * len(self.adj_list)
        self.components = []
        self.cycle = False
        self.prev = [None] * len(adj_list)
        self.tree_edges = [[] for _ in range(len(self.adj_list))]
        self.top_sort = []
        # self.tree_matrix = [[0] * len(self.adj_list) for _ in range(len(self.adj_list))]
        # self.matrix = [[0] * len(self.adj_list) for _ in range(len(self.adj_list))]

    # def from_list_to_matrix(self):
    #    for i in range(len(self.adj_list)):
    #        for j in self.adj_list[i]:
    #            self.matrix[i][j] = 1
    #            self.tree_matrix[i][j] = 1

    def print_graph_list(self):
        for i in self.adj_list:
            print(i)

    def print_edges_by_matrix(self, matrix):
        temp_matrix = matrix
        for i in range(len(temp_matrix)):
            for j in range(len(temp_matrix[i])):
                if temp_matrix[i][j] == 1:
                    print(i, j)
                    temp_matrix[j][i] = 0

    def print_edges_by_list(self, lst):
        set_edges = set()
        for i in range(len(lst)):
            for j in lst[i]:
                if (i, j) not in set_edges and (j, i) not in set_edges:
                    set_edges.add((i, j))
                    print(i, j)

    def dfs(self, point, curr_comp):
        self.colors[point] = self.gray
        for neib in self.adj_list[point]:
            if self.colors[neib] == self.white:
                self.tree_edges[point].append(neib)
                curr_comp.append(neib)
                self.prev[neib] = point
                self.dfs(neib, curr_comp)
            elif self.colors[neib] == self.gray:
                if neib != self.prev[point]:
                    self.cycle = True
                    # self.tree_matrix[neib][point] = 0
                    # self.tree_matrix[point][neib] = 0
                    # self.adj_list[point].remove(neib)
                    # self.adj_list[neib].remove(point)
                else:
                    self.tree_edges[point].append(neib)
                self.cycle = True

        self.colors[point] = self.black
        self.top_sort.append(point)
        if len(self.components) > 0 and curr_comp[0] != self.components[-1][0]:
            self.components.append(curr_comp)
        elif len(self.components) == 0:
            self.components.append(curr_comp)

    def count_components(self):
        for i in range(1, len(self.adj_list)):
            if self.colors[i] == self.white:
                self.dfs(i, [i])
        return self.components

    def is_tree(self):
        for i in range(1, len(self.adj_list)):
            if self.colors[i] == self.white:
                self.dfs(i, [i])
                if self.cycle:
                    return self.cycle
        return self.cycle

    def make_tree(self):
        self.dfs(1, [1])
        # self.print_edges(self.tree_matrix)
        self.print_edges_by_list(self.tree_edges)

    def has_cycle(self):
        for i in range(1, len(self.adj_list)):
            if self.colors[i] == self.white:
                self.dfs(i, [i])
            if self.cycle:
                return 1
        return 0

    def make_top_sort(self):
        for i in range(1, len(self.adj_list)):
            if self.colors[i] == self.white:
                self.dfs(i, [i])

        return self.top_sort[::-1]


v, e = map(int, input().split())
lst = [[] for _ in range(v + 1)]
for _ in range(e):
    a, b = map(int, input().split())
    lst[a].append(b)
    # lst[b].append(a)

graph = Graph(lst)

if graph.has_cycle():
    print(-1)
else:
    print(*graph.make_top_sort())
```
